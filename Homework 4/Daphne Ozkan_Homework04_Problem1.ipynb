### Question 1

According to the paper, AlexNet consists of five convolutional layers followed by three fully connected layers. We can calculate the number of parameters for each layer in the following way and sum them up to find the total number of parameters: 

The first convolutional layer has:
- Filter size: $11 \times 11$
- Input channels: 3 (RGB)
- Output channels (filters): 96
- Stride: 4
- Padding: No padding (valid padding)

The formula for the number of parameters in a convolutional layer is:
$$
\text{Parameters} = (F \times F \times C_{\text{in}} + 1) \times C_{\text{out}}
$$
where $F$ is the filter size, $C_{\text{in}}$ is the number of input channels, and $C_{\text{out}}$ is the number of output channels (filters). Thus, for Conv1:
$$
\text{Parameters} = (11 \times 11 \times 3 + 1) \times 96 = 364 \times 96 = 34,944
$$
So, the total number of parameters for Conv1 is $34,944$.

The second convolutional layer has:
- Filter size: $5 \times 5$
- Input channels: 96
- Output channels (filters): 256
- Stride: 1
- Padding: No padding (valid padding)

Using the same formula:
$$
\text{Parameters} = (5 \times 5 \times 96 + 1) \times 256 = 2,401 \times 256 = 614,656
$$
So, the total number of parameters for Conv2 is $614,656$.

The third convolutional layer has:
- Filter size: $3 \times 3$
- Input channels: 256
- Output channels (filters): 384
- Stride: 1
- Padding: No padding (valid padding)

Using the same formula:
$$
\text{Parameters} = (3 \times 3 \times 256 + 1) \times 384 = 2,305 \times 384 = 885,120

The fourth convolutional layer has:
- Filter size: $3 \times 3$
- Input channels: 384

Using the same formula:
$$
\text{Parameters} = (3 \times 3 \times 384 + 1) \times 384 = 3,457 \times 384 = 1,327,488

The fifth convolutional layer has:
- Filter size: $3 \times 3$
- Input channels: 384
- Output channels (filters): 256
- Stride: 1
- Padding: No padding (valid padding)

Using the same formula:
$$
\text{Parameters} = (3 \times 3 \times 384 + 1) \times 256 = 3,457 \times 256 = 884,992
$$
So, the total number of parameters for Conv5 is $884,992$.

The fully connected layers are as follows:

FC1 has:
- Input size: $6 \times 6 \times 256 = 9,216$
- Output size: 4096

The number of parameters for FC1 is:
$$
\text{Parameters} = (9,216 + 1) \times 4096 = 9,217 \times 4096 = 37,752,832
$$
So, the total number of parameters for FC1 is $37,752,832$.

FC2 has:
- Input size: 4096
- Output size: 4096

The number of parameters for FC2 is:
$$
\text{Parameters} = (4096 + 1) \times 4096 = 4,097 \times 4096 = 16,781,312
$$
So, the total number of parameters for FC2 is $16,781,312$.

FC3 has:
- Input size: 4096
- Output size: 1000 (for 1000 classes)

The number of parameters for FC3 is:
$$
\text{Parameters} = (4096 + 1) \times 1000 = 4,097 \times 1000 = 4,097,000
$$
So, the total number of parameters for FC3 is $4,097,000$.

Now, summing up all the parameters from each layer:
$$
\text{Total Parameters} = 34,944 + 614,656 + 885,120 + 1,327,488 + 884,992 + 37,752,832 + 16,781,312 + 4,097,000 = 62,378,344
$$
Thus, the total number of parameters in AlexNet is $62,378,344$.

### Question 3

To show that a stack of $N$ convolution layers, each with a filter size of $F \times F$, has the same receptive field as a single convolution layer with a filter size of $(NF - N + 1) \times (NF - N + 1)$, we can first consider a single convolution layer.
For a convolution layer with a filter size of $F \times F$ and stride 1, the receptive field is simply the size of the filter itself. So, the receptive field for one convolution layer is:

$$
\text{Receptive Field (Single Layer)} = F \times F
$$

Now, when we stack $N$ convolution layers, each with a filter size of $F \times F$, the receptive field increases. This is because each additional layer "sees" a larger portion of the input. Specifically, each new layer increases the receptive field by $(F - 1)$ pixels in both the height and width.

Thus, after stacking $N$ layers, the total receptive field is:

$$
\text{Receptive Field (Stacked Layers)} = (N \times (F - 1)) + 1
$$

This equation shows how the receptive field grows as we add more layers to the network.

Finally, to match the receptive field of $N$ stacked layers, we can use a single large convolution layer with a larger filter size. The receptive field of this single layer will be equivalent to the stacked layers and can be written in the following way:

$$
\text{Receptive Field (Single Large Layer)} = (N \times F - N + 1) \times (N \times F - N + 1)
$$

This shows that stacking $N$ convolution layers with a filter size of $F \times F$ results in the same receptive field as a single convolution layer with a receptive field of size $(NF - N + 1) \times (NF - N + 1)$.


To calculate the receptive field for 3 convolution layers, each with a filter size of $5 \times 5$, we substitute $N = 3$ and $F = 5$ into the formula:

$$
\text{Receptive Field} = (3 \times 5 - 3 + 1) \times (3 \times 5 - 3 + 1)
$$

Simplifying the expression:

$$
\text{Receptive Field} = (15 - 3 + 1) \times (15 - 3 + 1)
$$

$$
\text{Receptive Field} = (13) \times (13) = 169
$$

Thus, the receptive field for 3 convolution layers with a filter size of $5 \times 5$ is $169 \times 169$.

This shows that stacking smaller filters like $3 \times 3$ in VGG networks can result in a large receptive field, similar to the receptive field of architectures like AlexNet, which use larger filters.

The Inception module is designed to allow the network to learn a variety of features by using parallel convolutions with different filter sizes (such as $1 \times 1$, $3 \times 3$, and $5 \times 5$). This parallel approach enables the network to capture information at different scales and levels of abstraction simultaneously. The outputs from each of these parallel convolution operations are then concatenated to form a richer feature map.

The idea behind this architecture is to allow the model to simultaneously capture fine-grained details (with smaller filters like $1 \times 1$) and larger contextual information (with larger filters like $5 \times 5$) while maintaining computational efficiency. In some versions, such as the dimensionality reduction version, the network uses $1 \times 1$ convolutions to reduce the number of input channels before applying larger convolutions, which decreases the computational cost.

Thus, the Inception module helps the network capture diverse features at multiple scales, which can improve its performance across a wide range of tasks.

### Question 5

### (a)

The key difference between Fast R-CNN and Faster R-CNN, which allows Faster R-CNN to detect objects faster, is the introduction of the Region Proposal Network (RPN).

In Fast R-CNN, external methods like Selective Search are used to generate region proposals, which are then passed to the CNN for classification and bounding box regression. This extra step of generating proposals separately adds significant computational cost and slows down the detection process.

On the other hand, Faster R-CNN integrates the RPN directly into the network. The RPN generates region proposals as part of the network's forward pass, using shared convolutional layers with the detection network. This means that Faster R-CNN doesn't rely on external algorithms for proposal generation, making the process much more efficient and faster. Additionally, Faster R-CNN can be trained end-to-end, further streamlining the detection pipeline.

This integration of proposal generation with object detection is what makes Faster R-CNN faster than Fast R-CNN.

### (b)

A Region Proposal Network (RPN) is a neural network used to generate region proposals for object detection tasks. It was introduced in the Faster R-CNN architecture to replace traditional methods like Selective Search, which were computationally expensive and time-consuming.

The RPN works by sliding a small convolutional network over the feature map produced by the backbone network (like VGG or ResNet). At each sliding window location, the RPN predicts multiple region proposals, each represented by a bounding box with associated objectness scores (indicating how likely the box contains an object).

The RPN architecture consists of the following key components:
- **Convolutional layers**: These are used to extract features from the input image. The RPN typically uses a 3x3 convolution filter for this purpose.
- **Anchor boxes**: At each sliding window location, a set of predefined anchors with different aspect ratios and scales are used. These anchors serve as reference bounding boxes for the network to adjust.
- **Bounding box regression**: For each anchor, the RPN predicts the adjustments (delta values) to the anchor box to refine its location.
- **Objectness score**: The RPN also predicts a binary classification score for each anchor, indicating whether it contains an object or background.


### (c)

To explain how region proposals are generated from RPN, we can consider the example image below, which is the Mona Lisa.

1. **Input Image**: The input image is passed through a convolutional network (like VGG or ResNet) to extract feature maps. Let's assume the resulting feature map is of size $32 \times 32 \times 512$ which is a smaller resolution but rich in information.

2. **Sliding Window**: The RPN applies a small convolutional filter (typically $3 \times 3$) over the feature map. The filter slides over the feature map, generating region proposals at each location. For each sliding window location, the RPN generates a set of predefined anchor boxes (e.g., 9 different anchors with different aspect ratios and scales).

3. **Anchor Boxes**: The anchor boxes are predefined at each sliding window location. These anchors have different aspect ratios, such as $1:1$, $2:1$, and $1:2$, and different scales. For example, one anchor may have a square shape, while another may have a rectangular shape to match different object structures.

4. **Bounding Box Regression**: For each anchor, the RPN predicts adjustments (delta values) to refine the bounding box to better fit the object. For example, the RPN may adjust the position of an anchor box around Mona Lisa’s face to make the bounding box more accurate.

5. **Objectness Score**: The RPN also predicts an objectness score for each anchor, indicating the likelihood that the anchor contains an object (Mona Lisa’s face, hands, or background) rather than being background.

6. **Proposals**: The region proposals are generated by applying the predicted bounding box adjustments to the anchors. Proposals with a high objectness score (indicating that they likely contain the object of interest, in this case, Mona Lisa’s face or hands) are kept. Proposals with low objectness scores are discarded.

To summarize, the RPN uses predefined anchor boxes at each location on the feature map, predicts bounding box adjustments, and assigns objectness scores to generate region proposals. In this case, the proposals would correspond to regions around Mona Lisa’s face, hands, and other significant parts of the image.


### (d)

To reduce the large number of overlapping proposals generated by the Region Proposal Network (RPN), Faster R-CNN uses a technique called Non-Maximum Suppression (NMS).

We can use the same example of the Mona Lisa image above to explain how this technique works:

1. **Overlapping proposals**: After the RPN generates region proposals for the Mona Lisa image, there will be many proposals that overlap, especially around key objects like her face and hands. Since the RPN produces a large number of candidate regions, many of these proposals may be redundant or highly similar, covering the same area.

2. **Sorting proposals**: The proposals are first sorted by their objectness score (i.e., the probability that a given region contains an object). For example, proposals with high objectness scores that correspond to Mona Lisa's face or hands will be ranked higher than background regions.

3. **Suppressing overlapping proposals**: Starting from the highest-scoring proposal (e.g., the region around Mona Lisa’s face), the algorithm compares it to the other proposals. If two proposals have a high overlap (measured by Intersection over Union, or IoU), typically greater than 0.5, the one with the lower score is suppressed. This means that overlapping proposals around Mona Lisa's face or hands that are redundant will be removed.

4. **Final selection**: This process continues until only the best proposals (with minimal overlap) remain. The final set of proposals will contain regions that represent distinct parts of the image, such as Mona Lisa’s face, hands, and other relevant features, while filtering out background or redundant regions.

For example, if the RPN generates 10,000 proposals for the Mona Lisa image, NMS will remove those that overlap too much, leaving only about 2,000 high-quality, non-overlapping proposals for further classification and bounding box refinement.
