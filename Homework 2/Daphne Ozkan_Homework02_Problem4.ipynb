```python
import matplotlib.pyplot as plt

# define dataset points
class_1 = [(-1, -1), (1, 1), (-3, -3), (4, 4)]
class_2 = [(-1, 1), (1, -1), (-5, 2), (4, -8)]

# separate x and y coordinates for plotting
x1_class_1, x2_class_1 = zip(*class_1)
x1_class_2, x2_class_2 = zip(*class_2)

# plotting the data points
plt.figure(figsize=(6,6))
plt.scatter(x1_class_1, x2_class_1, color='blue', label='Class 1')
plt.scatter(x1_class_2, x2_class_2, color='red', label='Class 2')

plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
plt.title("Scatter Plot of Dataset Points")
plt.axhline(0, color='gray', linestyle='--', linewidth=0.5)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.grid()

plt.show()
```

Looking at the plot of the data points above, we can see that the dataset is not linearly separable, since we cannot draw a single straight line to separate blue points from red points. Because the dataset is not linearly separable, a linear classifier cannot be trained using $x_1$ and $x_2$ to classify this dataset. Instead, we would need a nonlinear classifier or feature transformation to make the dataset separable.

### Question 4

Nonlinear transformations are important in classification problems because they allow us to represent data in a way that makes it easier to separate different classes. Some datasets are not linearly separable in their original form, meaning no straight-line boundary can divide the classes. By applying a nonlinear transformation, we can map the data into a new space where a simple decision boundary, such as a threshold, can separate the classes correctly. This helps improve the accuracy of classification models and makes them more effective at capturing complex relationships in the data.
